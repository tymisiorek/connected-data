```python
all_board_members = []
for year in valid_years:
    boards_path = f"{path_temp_data}{year}_boards.csv"
    df = pd.read_csv(boards_path)
    all_board_members.append(df)

all_board_members_df = pd.concat(all_board_members, ignore_index=True)
all_board_members_df = all_board_members_df.drop_duplicates()

all_board_names = all_board_members_df["Name"].values
```

```python
billionaire_df = pd.read_csv(billionaires_path)
billionaire_names = billionaire_df["full_name"].values
billionaire_names = set([value.replace("& family", "") for value in billionaire_names])

```

```python
similarity_threshold = 90
billionaire_board_members = []

for member in all_board_names:
    result = process.extractOne(member, billionaire_names, score_cutoff=similarity_threshold)
    if result is not None:
        print("fad.nfnkadk")
        match, score, _ = result
        billionaire_board_members.append((member, match, score))


print(f"Number of board members who are billionaires (similarity â‰¥ {similarity_threshold}%): {len(billionaire_board_members)}")
for board_member, matched_billionaire, similarity in billionaire_board_members:
    print(f"Board Member: {board_member}, Matched Billionaire: {matched_billionaire}, Similarity: {similarity}%")
```

```python
billionaire_board_members_set = set(billionaire_board_members)
print(f"Number of board members who are billionaires (similarity â‰¥ {similarity_threshold}%): {len(billionaire_board_members_set)}")

for board_member, matched_billionaire, similarity in billionaire_board_members_set:
    print(f"Board Member: {board_member}, Matched Billionaire: {matched_billionaire}, Similarity: {similarity}%")
```

```python
from fuzzywuzzy import fuzz
unique_board_members = []

# Iterate through each board member in the original set
for board_member, matched_billionaire, similarity in billionaire_board_members_set:
    is_duplicate = False
    
    for unique_member, _, _ in unique_board_members:
        similarity_score = fuzz.ratio(board_member, unique_member)
        
        if similarity_score >= 97:
            is_duplicate = True
            break
    
    if not is_duplicate:
        unique_board_members.append((board_member, matched_billionaire, similarity))

# Output the cleaned list of board members
print(f"Unique board members: {unique_board_members}")
print(f"Length of unique board members list: {len(unique_board_members)}")
```
